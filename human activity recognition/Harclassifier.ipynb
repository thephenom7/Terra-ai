```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix
```

```python
X_train = pd.read_csv('/kaggle/input/trainhar/X_train.txt', delim_whitespace=True)
X_train=X_train.to_numpy()
```

```python
y_test.shape
```

```python
from datetime import datetime
from sklearn import metrics
from sklearn.model_selection import GridSearchCV

def train(model,X_train,y_train,X_test,y_test):
    train_start_time=datetime.now()
    print("training the model has begin...")
    model.fit(X_train,y_train)
    print("Done..")
    train_end_time=datetime.now()
    print(f'Training time is {train_end_time-train_start_time}')
    
    print("Predicting Test data ...")
    y_pred=model.predict(X_test)
    print("Done ..")
    
    accuracy = metrics.accuracy_score(y_true=y_test, y_pred=y_pred)
    print("--------------")
    print("Accuracy")
    print("---------------")
    print(accuracy)
    cm = metrics.confusion_matrix(y_test, y_pred)
    print("--------------")
    print("Confusion matrix")
    print("---------------")
    print(cm)
    
    return model
    
```

```python
from sklearn import linear_model
```

```python
parameters = {'C':[0.01, 0.1, 1, 10, 20, 30], 'penalty':['l2','l1']}
log_reg = linear_model.LogisticRegression()
log_reg_grid = GridSearchCV(log_reg, param_grid=parameters, cv=10, verbose=1, n_jobs=-1)
model=  train(log_reg_grid, X_train, y_train, X_test, y_test)
```

```python

print('--------------------------')
print('|      Best Estimator     |')
print('--------------------------')
print('\n\t{}\n'.format(model.best_estimator_))


    # parameters that gave best results while performing grid search
print('--------------------------')
print('|     Best parameters     |')
print('--------------------------')
print('\tParameters of best estimator : \n\n\t{}\n'.format(model.best_params_))


    #  number of cross validation splits
print('---------------------------------')
print('|   No of CrossValidation sets   |')
print('--------------------------------')
print('\n\tTotal numbre of cross validation sets: {}\n'.format(model.n_splits_))


    # Average cross validated score of the best estimator, from the Grid Search 
print('--------------------------')
print('|        Best Score       |')
print('--------------------------')
print('\n\tAverage Cross Validate scores of best estimator : \n\n\t{}\n'.format(model.best_score_))
```

```python
from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import GridSearchCV
from keras.models import Sequential
from keras.layers import Dense
def build_classifier(optimizer):
    classifier = Sequential()
    classifier.add(Dense(units = 512, kernel_initializer = 'uniform', activation = 'relu', input_dim =X_train.shape[1]))
    classifier.add(Dense(units = 1024, kernel_initializer = 'uniform', activation = 'relu'))
    classifier.add(Dense(units = 512, kernel_initializer = 'uniform', activation = 'relu'))
    classifier.add(Dense(units = 6, kernel_initializer = 'uniform', activation = 'softmax'))
    classifier.compile(optimizer = optimizer, loss = 'categorical_crossentropy', metrics = ['accuracy'])
    return classifier
classifier = KerasClassifier(build_fn = build_classifier)
parameters = {'batch_size': [32],
              'epochs': [25],
              'optimizer': ['adam']}
ann_search = GridSearchCV(estimator = classifier,
                           param_grid = parameters,
                           scoring = 'accuracy',cv=2)
print("Training has begin")
start_time=datetime.now()
ann_model = ann_search.fit(X_train, y_train)
end_time=datetime.now()
print("Done...")
print("Time for training ".format(end_time-start_time))
```

```python
print("Predicting Test data ...")
y_pred=ann_model.predict(X_test)
print("Done ..")
    
accuracy = metrics.accuracy_score(y_true=y_test, y_pred=y_pred)
print("--------------")
print("Accuracy")
print("---------------")
print(accuracy)
cm = metrics.confusion_matrix(y_test, y_pred)
print("--------------")
print("Confusion matrix")
print("---------------")
print(cm)
```
